# failureType의 유형별 빈도 분석 2

```python
df2=pd.read_pickle("/content/drive/MyDrive/ACK2025_Wafer/MIR_WM811K_Dataset/WM811K.pkl")
```

```python
# failureType 및 trainTestLabel을 숫자형으로 설정
df2['failureNum']=df2.failureType
df2['trainTestNum']=df2.trainTestLabel
mapping_type={'Center':0,'Donut':1,'Edge-Loc':2,'Edge-Ring':3,'Loc':4,'Random':5,'Scratch':6,'Near-full':7,'none':8}
mapping_traintest={'Training':0,'Test':1}
df2['failureNum'] = df2['failureType'].apply(lambda x: x[0] if isinstance(x, (list, np.ndarray)) and len(x)>0 else x)
df2['failureNum'] = df2['failureNum'].map(mapping_type)
df2['trainTestNum'] = df2['trainTestLabel'].apply(lambda x: x[0] if isinstance(x, (list, np.ndarray)) and len(x)>0 else x)
df2['trainTestNum'] = df2['trainTestNum'].map(mapping_traintest)
```

```python
df2.head(40)
```

**[0,0]은 NaN으로 설정되어 있음을 35번째 데이터를 통해 확인할 수 있다.**

```python
total_wafers = len(df2)
print('전체 웨이퍼 개수 :',total_wafers)
```

```python
# withlabel : label이 명확히 표시된 모든 데이터
df2_withlabel = df2[(df2['failureNum']>=0) & (df2['failureNum']<=8)]
df2_withlabel = df2_withlabel.reset_index()

# withpattern : label이 명확히 표시되고 결함 패턴이 있는 데이터
df2_withpattern = df2[(df2['failureNum']>=0) & (df2['failureNum']<=7)]
df2_withpattern = df2_withpattern.reset_index()

# nonpattern : label이 명확히 표시되고 결함 패턴이 없는 데이터
df2_nonpattern = df2[(df2['failureNum']==8)]

# 각각의 데이터의 빈도 출력
df2_withlabel.shape[0], df2_withpattern.shape[0], df2_nonpattern.shape[0]
```

```python
# 파이 분석 및 분포도 제작
import matplotlib.pyplot as plt
%matplotlib inline

from matplotlib import gridspec
fig = plt.figure(figsize=(20, 4.5))
gs = gridspec.GridSpec(1, 2, width_ratios=[1, 2.5])
ax1 = plt.subplot(gs[0])
ax2 = plt.subplot(gs[1])

no_wafers=[total_wafers-df2_withlabel.shape[0], df2_withpattern.shape[0], df2_nonpattern.shape[0]]

colors = ['silver', 'orange', 'gold']
explode = (0.1, 0, 0)  # explode 1st slice
labels = ['no-label','label&pattern','label&non-pattern']
ax1.pie(no_wafers, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=140)

uni_pattern=np.unique(df2_withpattern.failureNum, return_counts=True)
labels2 = ['','Center','Donut','Edge-Loc','Edge-Ring','Loc','Random','Scratch','Near-full']
ax2.bar(uni_pattern[0],uni_pattern[1]/df2_withpattern.shape[0], color='gold', align='center', alpha=0.9)
ax2.set_title("FailureType Distribution with Pattern")
ax2.set_ylabel("% of pattern wafers")
ax2.set_xticklabels(labels2)

plt.show()
```



*   no-label : failureType의 값이 [0,0]로 표시되어 있는 데이터

    *   해당 데이터는 Label이 불분명한 데이터라고 판단됨
    *   해당 데이터는 학습 및 분류로 사용되지 않을 예정


*   label&non-patern : failureType의 값이 none으로 표시되어 있는 데이터

    *   해당 데이터는 Label이 확실하게 존재하지만 특별한 결함이 없는 데이터
    *   결함 패턴 학습 및 분류에 있어서 정상&비정상 클래스 중 정상 클래스로 분류될 데이터

*   label&pattern : failureType의 값이 none과 [0,0]을 제외한 나머지 데이터
    *   해당 데이터는 Label이 확실하게 존재하며 특정 결함이 존재하는 데이터
    *   결함 패턴 학습 및 분류에 있어서 정상&비정상 클래스 중 비정상 클래스로 분류될 데이터




# failureType과 trainTestLabel의 값이 모두 [0,0]인 데이터 확인

```python
import numpy as np

# 함수 제작
def is_00(val):
    return (isinstance(val, (list, np.ndarray)) and np.array_equal(val, [0, 0])) or str(val) == '[0, 0]'

abnormal_both = df[
    df['failureType'].apply(is_00) &
    df['trainTestLabel'].apply(is_00)
]

print(f"failureType과 trainTestLabel의 값이 모두 '[0, 0]'인 데이터의 수: {len(abnormal_both)}")
```

***failureType의 값이 [0,0]인 데이터의 개수와 trainTestLabel의 값이 [0,0]인 데이터의 개수가 같다는 것을 확인할 수 있다.***
